# 01 Corpus Analysis — Atayal (Zeyuan Dialect)
This notebook performs basic statistical analysis on the Atayal text corpus.
## 分析结果摘要 (Chapter 1)
- 句子数：49
- 总词数：204
- 词汇表大小：76
- 平均句子长度：4.2 个词
- 最高频词：ga (19次), biru (15次), ku (10次)

```python
# 读取泰雅语语料
with open("1.pyu", "r", encoding="utf-8") as f:
    lines = f.readlines()

# 去掉每行末尾的换行符
lines = [line.strip() for line in lines]

# 去掉空行
lines = [line for line in lines if line]

print(f"总句子数：{len(lines)}")
print(f"前3句：")
for line in lines[:3]:
    print(line)
```

```python
import string

sentence = "psba biru, lawkah su ga?"

# string.punctuation 包含所有标点
# 但我们要把 ' 从里面移除，因为泰雅语需要它
punctuation_to_remove = string.punctuation.replace("'", "")

# 去掉标点
cleaned = sentence.translate(str.maketrans("", "", punctuation_to_remove))

print(cleaned)
```

```python
import string
from collections import Counter

# 去掉标点（保留撇号）
punctuation_to_remove = string.punctuation.replace("'", "")

# 处理所有句子，分割成词
all_words = []
for line in lines:
    cleaned = line.translate(str.maketrans("", "", punctuation_to_remove))
    words = cleaned.lower().split()
    all_words.extend(words)

# 统计
word_counts = Counter(all_words)

print(f"总词数（含重复）：{len(all_words)}")
print(f"词汇表大小（不重复）：{len(word_counts)}")
print(f"\n最常见的10个词：")
for word, count in word_counts.most_common(10):
    print(f"  {word}: {count}")
```

```python
# 每句话的词数
sentence_lengths = [len(line.translate(
    str.maketrans("", "", punctuation_to_remove)).lower().split())
    for line in lines]

avg_length = sum(sentence_lengths) / len(sentence_lengths)
print(f"平均句子长度：{avg_length:.1f} 个词")
print(f"最短句子：{min(sentence_lengths)} 个词")
print(f"最长句子：{max(sentence_lengths)} 个词")
```
